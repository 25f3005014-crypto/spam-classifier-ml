```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Sample knowledge base
data = {
    "question": [
        "what is python",
        "what is machine learning",
        "what is regression",
        "what is classification",
        "what is overfitting",
        "what is ai",
        "what is numpy",
        "what is pandas",
        "how to start coding",
        "i dont know coding",
        "what should i learn first in programming",
        "how to learn python",
        "is coding hard",
        "how many hours should i study coding",
        "what is coding",
        "how to practice coding",
        "best way to learn programming",
        "can beginners learn python",
        "what is the easiest programming language",
        "how long does it take to learn coding",
        "how to improve coding skills",
        "what should i do if i dont understand coding",
        "how to debug errors",
        "how to write a program",
        "what is syntax",
        "why is my code not working",
        "how to learn faster",
        "how to stay motivated to study",
        "what is python",
        "what is machine learning",
        "what is regression",
        "what is classification",
        "what is overfitting",
        "what is ai",
        "what is numpy",
        "what is pandas",
        "what is deep learning",
        "what is data science",
        "what is supervised learning",
        "what is unsupervised learning",
        "what is clustering",
        "what is a dataset",
        "what is a model",
        "what is training data",
        "what is testing data",
        "what is accuracy",
        "what is precision",
        "what is recall",
        "what is f1 score",
        "what is normalization",
        "what is standardization",
        "what is gradient descent",
        "what is a neural network",
        "what is tensorflow",
        "what is scikit learn",
        "what is matplotlib",
        "what is feature engineering",
        "what is natural language processing",
        "what is nlp",
        "what is chatbot",
        "what is cosine similarity",
        "what is tf idf",
        "what is data visualization",
        "what is big data",
        "what is cloud computing",
        "what is api",
        "what is flask",
        "what is streamlit",
        "what is version control",
        "what is github",
        "what is git",
        "what is debugging",
        "what is an algorithm",
        "what is a variable",
        "what is a loop",
        "what is a function",
        "what is object oriented programming",
        "what is class in python",
        "what is inheritance",
        "what is recursion",
        "what is list in python",
        "what is dictionary in python",
        "what is tuple in python"
    ],
    "answer": [
        "Python is a programming language used for web, data science, and AI.",
        "Machine learning is a subset of AI that allows systems to learn from data.",
        "Regression is used to predict continuous values.",
        "Classification is used to categorize data into classes.",
        "Overfitting occurs when a model learns training data too well.",
        "Artificial Intelligence is the simulation of human intelligence in machines.",
        "NumPy is a library used for numerical computations in Python.",
        "Pandas is a library used for data manipulation and analysis.",
        "You can start coding by learning basic concepts like variables, loops, and functions. Python is a great beginner-friendly language.",
        "That's okay! Everyone starts from zero. Begin with simple topics and practice regularly.",
        "You should start with basics like variables, loops, conditions, and functions.",
        "You can learn Python by practicing daily, watching tutorials, and solving small problems.",
        "Coding may seem hard at first, but with practice it becomes easier.",
        "Start with 1–2 hours daily and focus on consistency rather than long hours.",
        "Coding is writing instructions for a computer to perform tasks.",
        "You can practice coding by solving problems and building small projects.",
        "The best way to learn programming is by practicing and building projects.",
        "Yes, Python is one of the easiest languages for beginners.",
        "Python is considered one of the easiest programming languages.",
        "It depends on practice, but basics can be learned in a few weeks.",
        "Practice regularly, build projects, and learn from mistakes.",
        "If you don't understand something, break it into smaller parts and try again.",
        "Debugging means finding and fixing errors in your code.",
        "A program is written by combining logic with programming syntax step by step.",
        "Syntax refers to the rules of writing code in a programming language.",
        "Check for syntax errors, missing brackets, or incorrect logic.",
        "Stay consistent, practice daily, and revise concepts regularly.",
        "Set small goals, track progress, and remember why you started.",

        "Python is a programming language used for web development, data science, and AI.",
        "Machine learning is a subset of AI that allows systems to learn from data.",
        "Regression is used to predict continuous numerical values.",
        "Classification is used to categorize data into different classes.",
        "Overfitting occurs when a model learns training data too well and performs poorly on new data.",
        "Artificial Intelligence is the simulation of human intelligence in machines.",
        "NumPy is a Python library used for numerical computations.",
        "Pandas is a Python library used for data analysis and manipulation.",
        "Deep learning is a subset of machine learning that uses neural networks with many layers.",
        "Data science involves extracting insights from data using statistics and machine learning.",
        "Supervised learning uses labeled data to train models.",
        "Unsupervised learning finds patterns in data without labeled outputs.",
        "Clustering is a technique used to group similar data points together.",
        "A dataset is a collection of data used for training and testing models.",
        "A model is a mathematical representation used to make predictions.",
        "Training data is used to train the machine learning model.",
        "Testing data is used to evaluate the performance of a model.",
        "Accuracy measures how many predictions are correct.",
        "Precision measures how many predicted positives are actually correct.",
        "Recall measures how many actual positives are correctly predicted.",
        "F1 score is the harmonic mean of precision and recall.",
        "Normalization scales data between a fixed range, usually 0 and 1.",
        "Standardization transforms data to have mean 0 and standard deviation 1.",
        "Gradient descent is an optimization algorithm used to minimize errors.",
        "A neural network is a model inspired by the human brain used in deep learning.",
        "TensorFlow is a popular library for deep learning developed by Google.",
        "Scikit-learn is a Python library for machine learning.",
        "Matplotlib is a library used for plotting graphs in Python.",
        "Feature engineering is the process of selecting and transforming variables for a model.",
        "Natural Language Processing is a field of AI that deals with text and language.",
        "NLP stands for Natural Language Processing.",
        "A chatbot is a program that can simulate conversation with users.",
        "Cosine similarity measures similarity between two text vectors.",
        "TF-IDF is a technique to convert text into numerical values based on importance.",
        "Data visualization is the graphical representation of data.",
        "Big data refers to extremely large datasets that require special tools to process.",
        "Cloud computing is the delivery of computing services over the internet.",
        "An API allows communication between different software systems.",
        "Flask is a lightweight Python web framework.",
        "Streamlit is a framework to build data apps easily.",
        "Version control is used to track changes in code.",
        "GitHub is a platform to host and manage code repositories.",
        "Git is a version control system.",
        "Debugging is the process of fixing errors in code.",
        "An algorithm is a step-by-step procedure to solve a problem.",
        "A variable is used to store data in a program.",
        "A loop is used to repeat a block of code.",
        "A function is a reusable block of code.",
        "Object Oriented Programming is a programming paradigm based on objects and classes.",
        "A class in Python is a blueprint for creating objects.",
        "Inheritance allows one class to inherit properties from another.",
        "Recursion is a function that calls itself.",
        "A list is a collection of ordered elements in Python.",
        "A dictionary stores data in key-value pairs.",
        "A tuple is an immutable collection of elements."
    ]
}

df = pd.DataFrame(data)

# Vectorize questions
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df["question"])

# Chat function
def chatbot(user_input):
    user_vec = vectorizer.transform([user_input])
    similarity = cosine_similarity(user_vec, X)
    index = similarity.argmax()

    if similarity[0][index] < 0.2:
        return "Sorry, I don't understand. Please ask something else."

    return df["answer"][index]

# Chat loop
print("🤖 AI Study Chatbot (type 'exit' to stop)")
while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("Chatbot: Goodbye! Keep learning 😊")
        break
    response = chatbot(user_input)
    print("Chatbot:", response)
```
